```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os.path
```

```python
concept = pd.read_csv('SP500_concepts.csv')
```

```python
concept = concept.sort_values(by=['Symbol']).reset_index(drop=True)
```

```python
N = 0
for i in range(concept.shape[0]):
    name = concept['Symbol'][i]
    if os.path.exists('./dataset/'+name+'.csv'):
        N +=1
        
N
```

```python
name_lst = []
for i in range(concept.shape[0]):
    name = concept['Symbol'][i]
    if os.path.exists('./dataset/'+name+'.csv'):
        name_lst.append(name)
```

```python
pre_concept = pd.DataFrame({'Symbol':name_lst})
pre_concept = pre_concept.merge(concept, how='inner', on='Symbol').drop('Security', axis=1)
pre_concept.columns=['Symbol', 'Sector', 'SubInd', 'Headloc', 'Date_added', 'CIK', 'Founded']
pre_concept
```

```python
N, K = pre_concept.shape
```

```python
relation = np.zeros((N, N, K-1))
for i in range(N):
    for j in range(N):
        for k in range(K-1):
            if pre_concept.iloc[i, k+1]==pre_concept.iloc[j, k+1]:
                relation[i, j, k] = 1
                
relation
```

```python
np.save('relation.npy', relation)
```
